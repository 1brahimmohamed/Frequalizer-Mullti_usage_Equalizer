```python
import librosa
from IPython.display import Audio as audioPlayer
```

```python
voice, sampling_rate = librosa.load("data/Sartaman.wav")
print(voice)
```

```python
audioPlayer(data=voice, rate=sampling_rate, autoplay=False)
```

```python
# n steps should be the slider value
edited_voice = librosa.effects.pitch_shift(voice, sr=sampling_rate, n_steps=-6.0)
audioPlayer(data=edited_voice, rate=sampling_rate, autoplay=False)
```
